* The info command confirmed that there is an inconsistency with the data type of date.
```python
#Grouping:
#1. What is the total number of confirmed cases, deaths, and recoveries for each country in the dataset?
total = group_data.groupby('Country/Region').sum()
conf = total[(['Confirmed', 'Deaths', 'Recovered'])].reset_index()
print('The total number of confirmed cases, deaths and recoveries are as show in the table below:')
conf
```

```python
conf.info()
```

*  The groupby command and the .sum() commnad were used to group and sum the confirmed cases, death and recovered covid-19 cases for each country under consideration. The .reset_index() command was used to reset the index of the dataframe.
*  The .info() command shows that the conf dataframe has 37 rows with 4 columns

```python
#2. How does the distribution of COVID-19 cases vary across different countries?
#Plotting a line graph to sbhow the distribution of covid-19 cases across different countries
Countries= conf['Country/Region']
cases= conf['Confirmed']
xticks_label= Countries
plt.figure(figsize=(10,6))
plt.plot(Countries, cases)
plt.title('Distribution of COVID-19 cases in different countries')
plt.xlabel('Coutries')
plt.ylabel('Confrimed COVID-19 cases')
plt.xticks(xticks_label, rotation=75)
plt.show()
```

*  The figure above is a line graph showing the distribution of confirmed covid-19 cases across different countries.
*  The gentle rise of confirmed cases Kuwait, show a of dropped cases in Kyrgyzstan and plateau from Krgyzstan to Maritius. 
*  A sudden rise in the number of cases reaching its peak in Mexcio was observed and the cases steeply dropping in Moldovia. 
*  The graph almost plateau from Monaco to Nepal and a rise in cases was observed in Netherlands and steeply dropping in New Zealand. 
*  The begun to rise again from North Macedonia and almost reached a competing peak with Mexico in Pakistan.

```python
#- Pivot Tables:
#1. Can you create a pivot table to show the total number of confirmed cases, deaths, and recoveries for each date and country?
#A pivot table showing total number of confirmed cases, deaths and recovered for each date and country
conf_deaths_recoveries= group_data.pivot_table(values = ['Confirmed', 'Deaths', 'Recovered'],
                                        index = 'Date',
                                        columns= 'Country/Region',
                                        aggfunc= 'sum')
conf_deaths_recoveries
```

```python
#Pivot table for only confirmed cases
con= group_data.pivot_table(values = ['Confirmed'],
                                        index = 'Date',
                                        columns= 'Country/Region',
                                        aggfunc= 'sum')
con
```

```python
#Pivot table for only deaths
deaths= group_data.pivot_table(values = ['Deaths'],
                                        index = 'Date',
                                        columns= 'Country/Region',
                                        aggfunc= 'sum')
deaths
```

```python
#Pivot table for only recovered cases
recovered= group_data.pivot_table(values = ['Recovered'],
                                        index = 'Date',
                                        columns= 'Country/Region',
                                        aggfunc= 'sum')
recovered
```

*  The conf_deaths_recoveries dataframe is a pivot table showing the confirmed, deaths and recovered cases with date as index, columns as country/region and aggfunc as sum
*  Sub pivot tables for confirmed (con), death (deaths) and recovered cases (recovered) only were also created.

```python
#- Time Series Analysis:
#1. How has the number of confirmed cases changed over time for each country?
#2. Are there any noticeable trends or patterns in the data?
#Convert the date in group_data dataframe to datetime
group_data['Date'] = pd.to_datetime(group_data['Date'])
#setting figure size
plt.figure(figsize=(12, 6))
#ploting date on the x-axis and confirmed cases pn the y-axis
plt.plot(group_data['Date'], group_data['Confirmed'])
#Adding axes labels
plt.title('Time series analysis of confirmed covid cases over time')
plt.xlabel('Time (date)')
plt.ylabel('Number of confirmed covid cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

*  The time series analysis shows the trend of how the confirmed covid-19 changed over time
*  The graph shows that the infection of covid-19 was small (almost zero) at the beginning of 2020 but by April, 2020 the infections begun rise reaching its peak by July, 2020 ending.

```python
#- Plotting with Matplotlib/Seaborn:
#1. Can you create a line plot showing the trend of confirmed cases over time for a specific country?
Kuwait= ['Kuwait']
Kuwait_cases = covid_data[covid_data['Country/Region'].isin(Kuwait)]
#Kuwait_cases
Kuwait_cases['Date']= pd.to_datetime(Kuwait_cases['Date'])
plt.figure(figsize=(10,6))
plt.plot(Kuwait_cases['Date'], Kuwait_cases['Confirmed'])
plt.title('Confirmed covid-19 cases in Kuwait over time')
plt.xlabel('Time (in date)')
plt.ylabel('Number of confirmed covid-19 cases in Kuwait')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

*  This graph shows the trend of the cahnge in confirmed covid-19 cases in Kuwait over time 
*  The infection were almost zero at the beginning of 2020 and begun to rise from mid April and until it reached its peak at the end of July, 2020.
*  The trend of the graph for infections of covid-19 has almost the same shape as the global trend of covid-19 confirmed cases

```python
#2. Can you perform any additional analyses or visualizations that provide insights into the COVID-19 data for your assigned countries?
#Plot a pie chart the continuous variables
label= conf['Country/Region']
plt.figure(figsize=(12, 10))
plt.pie(conf['Confirmed'], autopct='%1.1f%%')
plt.title('Distribution of confirmed covid-19 cases in various countries')
plt.legend(label, labels= label, bbox_to_anchor=(1, 1))
plt.show()
```

```python
#Pie chart plot to show the proportions of WHO Regions that were affected with Covid-19
WHO_Reg= covid_data['WHO Region'].value_counts()
WHO_Reg
labels = ['Europe', 'Africa', 'Americas', 'Eastern Mediterranean', 'Western Pacific', 'South-East Asia']
plt.figure(figsize=(10,6))
plt.pie(WHO_Reg, labels= labels, autopct='%1.1f%%', startangle=90)
plt.title('Disitribution of WHO Regions that were affected with covid-19')
plt.show()
```

*  The charts shows that Europe was the worse hit with covid-19 cosntituting 29.9% followed by Africa constituting 25.7% and South-East Asia was the least hit constituting 5.3%
